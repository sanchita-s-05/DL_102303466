```python
import numpy as np
import pandas as pd

data = pd.read_csv("/content/abalone.data", header=None)

data.columns = ["Sex","Length","Diameter","Height","Whole_weight",
                "Shucked_weight","Viscera_weight","Shell_weight","Rings"]

data["Sex"] = data["Sex"].map({"M":0,"F":1,"I":2})

X = data.drop("Rings",axis=1).values
y = data["Rings"].values.reshape(-1,1)

X = (X - X.mean(axis=0)) / X.std(axis=0)

n_inputs = X.shape[1]
n_hidden = 10
n_outputs = 1

W1 = np.random.randn(n_inputs,n_hidden)
b1 = np.zeros((1,n_hidden))

W2 = np.random.randn(n_hidden,n_outputs)
b2 = np.zeros((1,n_outputs))

lr = 0.001
epochs = 500

def relu(x):
    return np.maximum(0,x)

def relu_derivative(x):
    return x > 0

for epoch in range(epochs):

    Z1 = np.dot(X,W1) + b1
    A1 = relu(Z1)

    Z2 = np.dot(A1,W2) + b2
    y_pred = Z2

    loss = np.mean((y_pred - y)**2)

    dZ2 = y_pred - y
    dW2 = np.dot(A1.T,dZ2)/len(X)
    db2 = np.sum(dZ2,axis=0,keepdims=True)/len(X)

    dA1 = np.dot(dZ2,W2.T)
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = np.dot(X.T,dZ1)/len(X)
    db1 = np.sum(dZ1,axis=0,keepdims=True)/len(X)

    W2 -= lr*dW2
    b2 -= lr*db2
    W1 -= lr*dW1
    b1 -= lr*db1

    if epoch % 50 == 0:
        print("Epoch:",epoch,"Loss:",loss)

pred = np.dot(relu(np.dot(X,W1)+b1),W2)+b2

print("Predicted vs Actual")
for i in range(10):
    print(pred[i][0],y[i][0])
```
