```python
# Step 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load Dataset
data = pd.read_csv('multiple_linear_regression_dataset.csv')

print("Dataset Preview:")
print(data.head())


# Step 3: Separate features and target
X = data[['age','experience']].values
y = data['income'].values


# Step 4: Normalize features (important for gradient descent)
X = (X - X.mean(axis=0)) / X.std(axis=0)


# Step 5: Initialize weights and bias
weights = np.zeros(X.shape[1])
bias = 0

learning_rate = 0.01
epochs = 1000
n = len(y)


# Step 6: Training using Gradient Descent (Perceptron style)
for epoch in range(epochs):

    # predictions
    y_pred = np.dot(X, weights) + bias

    # error
    error = y_pred - y

    # gradients
    dw = (1/n) * np.dot(X.T, error)
    db = (1/n) * np.sum(error)

    # update weights
    weights = weights - learning_rate * dw
    bias = bias - learning_rate * db


# Step 7: Final model parameters
print("\nLearned Weights:", weights)
print("Bias:", bias)


# Step 8: Predictions
predictions = np.dot(X, weights) + bias

print("\nActual vs Predicted")
for i in range(len(y)):
    print("Actual:", y[i], " Predicted:", round(predictions[i],2))


# Step 9: Model Accuracy using R2 score
ss_total = np.sum((y - y.mean())**2)
ss_residual = np.sum((y - predictions)**2)

r2 = 1 - (ss_residual/ss_total)

print("\nR2 Score:", r2)
```
