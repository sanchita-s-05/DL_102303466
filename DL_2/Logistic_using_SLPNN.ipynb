```python
import numpy as np
import pandas as pd

data = pd.read_csv('/content/glass.csv')

X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values


def train_test_split(X, y, test_size=0.2):
    n_samples = X.shape[0]
    split_index = int(n_samples * (1 - test_size))

    indices = np.random.permutation(n_samples)
    X, y = X[indices], y[indices]

    X_train = X[:split_index]
    X_test = X[split_index:]
    y_train = y[:split_index]
    y_test = y[split_index:]

    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = train_test_split(X, y)

class LogisticRegression:

    def __init__(self, lr=0.1, epochs=100):
        self.lr = lr
        self.epochs = epochs
        self.w = None
        self.E = []
        self.w_list = []
        self.outputLayer = 1

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def feedforward(self, X):
        z = np.dot(X, self.w)
        h = self.sigmoid(z)
        return h

    def error(self, h, y):
        error = (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
        self.E.append(error)
        return error

    def backprop(self, X, y, h):
        self.delta_E_w = np.dot(X.T, (h - y)) / X.shape[0]
        self.w_list.append(self.w.copy())
        self.w = self.w - self.lr * self.delta_E_w

    def fit(self, X, y):
        self.w = np.zeros(X.shape[1])
        for _ in range(self.epochs):
            h = self.feedforward(X)
            self.error(h, y)
            self.backprop(X, y, h)

    def predict(self, X):
        h = self.feedforward(X)
        return np.where(h >= 0.5, 1, 0)

    def accuracy(self, y_true, y_pred):
        return np.mean(y_true == y_pred)

model = LogisticRegression(lr=0.01, epochs=10000)
model.fit(X_train, y_train)

y_test_pred = model.predict(X_test)
test_accuracy = model.accuracy(y_test, y_test_pred)

print("Test Accuracy:", test_accuracy)
```
