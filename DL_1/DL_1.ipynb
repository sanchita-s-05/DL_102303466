```python
import numpy as np

# Step activation function
def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0


# Perceptron Training Function
def train_perceptron(X, y, lr=0.1, epochs=10):

    # initialize weights and bias
    w = np.zeros(X.shape[1])
    b = 0

    for epoch in range(epochs):
        for i in range(len(X)):

            net_input = np.dot(X[i], w) + b
            y_pred = step_function(net_input)

            error = y[i] - y_pred

            # weight update rule
            w = w + lr * error * X[i]
            b = b + lr * error

    return w, b


# Prediction function
def predict(X, w, b):
    results = []
    for x in X:
        net = np.dot(x, w) + b
        results.append(step_function(net))
    return results


# Input dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])


# Outputs for different gates
gates = {
    "AND": np.array([0,0,0,1]),
    "OR": np.array([0,1,1,1]),
    "NAND": np.array([1,1,1,0]),
    "NOR": np.array([1,0,0,0])
}


# Training and Testing
for gate in gates:

    print("\n------",gate,"Gate------")

    y = gates[gate]

    w, b = train_perceptron(X, y)

    print("Weights:", w)
    print("Bias:", b)

    predictions = predict(X, w, b)

    for i in range(len(X)):
        print(X[i], "->", predictions[i])
```
